```python
import re

class Part:
    #_number = 
    #_text = 
    #_answer = 
    
    def __init__(self, text, num):
        self._number = num
        if (text):
            text = re.split(r'-\nвлечет|-\nвлекут', text)
            self._text = text[0]
            if (len(text)>1):
                self._answer = text[1]
            else:
                self._answer = None
        else: 
            self._text, self._answer = None, None
    def genAnswer(self):
        pass
    
    def __str__ (self):
        return (str(self._number) + " -> " + self._text + "->" + self._answer)


class Article:
    #_chapter =
    #_number =
    #_subnumber =
    #_title =
    #_parts = 
    
    def __init__(self, text):
        number_of_article_re = re.compile('\d+\.\d+\.\d+-\d+\. |\d+\.\d+\.\d+\. |\d+\.\d+\. ')
        number_of_article = number_of_article_re.match(text).group(0)
        text = re.sub(number_of_article, '', text)
        
        self._title = re.match(r'.+', text).group(0)
        text = re.sub(self._title, '', text)
        self._title = self._title[:-1]
        
        number_of_article = re.split(r'\.', number_of_article)       
        number_of_article = number_of_article[:-1]
        
        self._chapter = number_of_article[0]
        self._number = number_of_article[1]
        if (len(number_of_article)>2):
            self._subnumber = number_of_article[2]
        else:
            self._subnumber = None
        
        text = re.sub(r'\(.+Федерал.+\)', '', text)
        text = re.sub(r'\s\s+', '\n', text)
        text = re.sub(r'^\n', '', text)
        self._setParts(text)
        #print (type(self._parts))
        
    def _setParts(self, text):
        self._parts = []
        parts = re.split(r'^\d+. |^\d+.\d+ |\n\d+. |\n\d+.\d+' , text)
        if (len(parts) > 1):
            for i, part in enumerate (parts[1:]):
                self._parts.append(Part(part, i+1))
        else:
            self._parts.append(Part(parts[0], 0))
        
    
    def __str__ (self):
        return (self._parts)
    
    def genText(self):
        pass

#import utils


doc_list = []
list_of_articles = part2.split('Статья ')

for article in list_of_articles:
    if (article):
        doc_list.append(Article(article))
```

```python

data = []

for doc in doc_list:
    tmpparts = []
    for part in doc._parts:
        tmpparts.append((part._number, part._text, part._answer))
    data.append([doc._chapter, doc._number, doc._subnumber, doc._title, tmpparts])
    
import pickle

with open("Dump.txt", "wb") as fp:
    pickle.dump(data, fp)
```

```python
from nltk.tokenize import RegexpTokenizer
def get_words(text):
    tokenizer = RegexpTokenizer("\w[^!.;,?\-\s]*\w|\w")
    s = []
    s = tokenizer.tokenize(text)
    return s
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem.snowball import RussianStemmer
import pymorphy2

def get_tokens(words):
    stop_words = stopwords.words("russian")
    stem = RussianStemmer()
    #lemma = WordNetLemmatizer()
    tokens = []
    for i in words:
        if (i not in stop_words):
            tokens.append(stem.stem(i.lower()))
    return tokens
```

```python
def token_list(dict_of_docs):
    out_list = []
    s = ' '
    for key, value in dict_of_docs.items():
        out_list.append(s.join(get_tokens(get_words(value[0]))))
    return out_list
        
```
